```python
sentence=input("Enter a sentence: ").lower()
vowels="aeiou"
v=c=0
for ch in sentence:
    if ch.isalpha():
        if ch in vowels:
            v+=1
        else:
            c+=1
print("Vowels:", v, "Consonants:", c)
```

```python
balance=1000
while True:
    print("1.Check Balance 2.Deposit 3.Withdraw 4.Exit")
    choice=input("Enter choice: ")
    if choice=='1':
        print("Balance:", balance)
    elif choice=='2':
        amt=int(input("Enter amount: "))
        balance+=amt
    elif choice=='3':
        amt=int(input("Enter amount: "))
        if amt<=balance:
            balance-=amt
        else:
            print("Insufficient funds")
    elif choice=='4':
        break
```

```python
def factorial(n):
    f=1
    for i in range(1,n+1):
        f*=i
    return f
print(factorial(5))
```

```python
def is_prime(n):
    if n<2:return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:return False
    return True
print(is_prime(17))
```

```python
def c_to_f(c):return (c*9/5)+32
def f_to_c(f):return (f-32)*5/9

print(c_to_f(30), "F")
print(f_to_c(86), "C")
```

```python
contacts = {}
while True:
    print("1.Add 2.Search 3.Delete 4.Exit")
    ch = input("Choice: ")
    if ch == '1':
        name = input("Name: ")
        num = input("Number: ")
        contacts[name] = num
    elif ch == '2':
        name = input("Name: ")
        print("Number:", contacts.get(name, "Not found"))
    elif ch == '3':
        name = input("Name: ")
        contacts.pop(name, None)
    elif ch == '4':
        break
```

```python
library=[]
while True:
    print("1.Add 2.Issue 3.Return 4.Exit")
    ch=input("Choice: ")
    if ch=='1':
        book=input("Book name: ")
        library.append(book)
    elif ch=='2':
        book=input("Book to issue: ")
        if book in library: library.remove(book)
    elif ch=='3':
        book=input("Book to return: ")
        library.append(book)
    elif ch=='4':
        break
```

```python
menu = {"Pizza":200, "Burger":100, "Pasta":150}
order = []
while True:
    item = input("Order item or 'done': ")
    if item.lower() == 'done':
        break
    if item in menu:
        order.append(menu[item])
print("Total Bill:", sum(order))
```
